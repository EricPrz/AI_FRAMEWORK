### Reading MNIST

```python
import pandas as pd
import numpy as np
import torch
from torch.utils.data import DataLoader, TensorDataset

# Load the EMNIST dataset
train = pd.read_csv("../data/emnist-balanced-train.csv")
test = pd.read_csv("../data/emnist-balanced-test.csv")

# Reestructuring: separate labels and images
train_label, train_image = train[train.columns[0]].to_numpy(), train[train.columns[1:]].to_numpy().reshape((-1, 1, 28, 28))/255
test_label, test_image = test[test.columns[0]].to_numpy(), test[test.columns[1:]].to_numpy().reshape((-1, 1, 28, 28))/255

# One-hot encoding for labels (optional for PyTorch, we can directly use integer labels for classification)
# But since you might want to keep one-hot encoding for some reason, we'll proceed with that.
train_label_one_hot = np.zeros((len(train_label), 47))
for label in range(train_label.shape[0]):
    train_label_one_hot[label, train_label[label]] = 1
train_label = train_label_one_hot

test_label_one_hot = np.zeros((len(test_label), 47))
for label in range(test_label.shape[0]):
    test_label_one_hot[label, test_label[label]] = 1
test_label = test_label_one_hot

# Convert data to PyTorch tensors
train_image = torch.tensor(train_image, dtype=torch.float32)  # Image tensor (float32)
train_label = torch.tensor(train_label, dtype=torch.float32)  # One-hot label tensor (float32)

test_image = torch.tensor(test_image, dtype=torch.float32)  # Image tensor (float32)
test_label = torch.tensor(test_label, dtype=torch.float32)  # One-hot label tensor (float32)

# Create DataLoader for batching during training and testing
batch_size = 128
train_dataset = TensorDataset(train_image, train_label)
test_dataset = TensorDataset(test_image, test_label)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Example of using DataLoader in training loop
# This replaces the manual batching process in the training and test functions
```

### Creating The Model

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        # First convolutional layer
        self.conv1 = nn.Conv2d(1, 6, kernel_size=5, stride=1, bias=False)
        self.relu1 = nn.ReLU()
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Second convolutional layer
        self.conv2 = nn.Conv2d(6, 16, kernel_size=5, stride=1, bias=False)
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # Batch normalization
        self.bn1 = nn.BatchNorm1d(256)  # Corrected for the fully connected layer's 1D input
        self.bn2 = nn.BatchNorm1d(84)

        # Fully connected layers
        self.lin1 = nn.Linear(256, 84, bias=False)
        self.lin2 = nn.Linear(84, 47, bias=False)

        # Dropout
        self.drop2 = nn.Dropout(0.2)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.maxpool1(x)
        
        x = self.conv2(x)
        x = self.maxpool2(x)
        
        # Flatten the output from the conv layers
        x = x.view(-1, 256)  # Flatten the tensor into shape (-1, 256)
        
        x = self.bn1(x)
        x = self.lin1(x)
        x = self.bn2(x)
        x = self.drop2(x)
        x = self.lin2(x)
        
        return x

# Create the model
model = Model()

# Set model to training mode
model.train()

# Loss function
loss_fn = nn.CrossEntropyLoss()

# Optimizer (Adadelta)
optimizer = optim.Adadelta(model.parameters())


```

#### Training The Model

```python
import torch
import torch.nn as nn
import torch.optim as optim

def train(model, x, y, batch_size, loss_fn, optim):
    # Convert x and y into PyTorch tensors
    x = torch.tensor(x, dtype=torch.float32)
    y = torch.tensor(y, dtype=torch.long)  # Assuming labels are integers (for classification)
    
    num_batches = len(y) // batch_size

    total_loss = 0
    total_correct = 0

    for batch in range(num_batches):
        # Get batch indices
        batch_start, batch_end = batch * batch_size, (batch + 1) * batch_size
        inputs = x[batch_start:batch_end]
        labels = y[batch_start:batch_end]
        
        # Zero the gradients
        optim.zero_grad()

        # Forward pass
        outputs = model(inputs)
        
        # Compute loss
        loss = loss_fn(outputs, labels)
        total_loss += loss.item()

        # Compute accuracy
        _, predicted = torch.max(outputs, 1)  # Get the predicted class
        correct = (predicted == labels).sum().item()
        total_correct += correct

        # Print the loss and progress
        if batch % (num_batches // 10) == 0:
            print(f"loss: {loss.item()} [{(batch+1)*batch_size}/{len(y)}]")

        # Backward pass
        loss.backward()
        optim.step()

    # Average loss and accuracy
    avg_loss = total_loss / num_batches
    avg_accuracy = total_correct / len(y)

    print(f"Avg loss: {avg_loss}")
    print(f"Train Accuracy: {avg_accuracy * 100:.2f}%\n")

    return avg_loss, avg_accuracy

```

```python
def test(model, x, y, batch_size=128):
    model.eval()  # Set the model to evaluation mode
    correct = 0
    num_batches = len(y) // batch_size

    # Convert x and y into PyTorch tensors
    x = torch.tensor(x, dtype=torch.float32)
    y = torch.tensor(y, dtype=torch.long)

    with torch.no_grad():  # Disable gradient computation to save memory and computation
        for batch in range(num_batches):
            # Get batch indices
            batch_start, batch_end = batch * batch_size, (batch + 1) * batch_size
            inputs = x[batch_start:batch_end]
            labels = y[batch_start:batch_end]

            # Forward pass
            outputs = model(inputs)

            # Get predicted classes
            _, predicted = torch.max(outputs, 1)

            # Calculate the number of correct predictions
            correct += (predicted == labels).sum().item()

    # Calculate accuracy
    accuracy = correct / len(y)

    print(f"Test Accuracy: {accuracy * 100:.2f}%\n")
    return accuracy
```

```python
import matplotlib.pyplot as plt

losses = []
train_precision = []

test_precision = []

epochs = 50

for epoch in range(epochs):
    print(f"Epoch {epoch + 1}\n-------------------------------")
    model.train()

    loss, precision = train(model, train_image, train_label, 128, loss_fn, optimizer)
    
    losses.append(loss)
    train_precision.append(precision)

    if epoch % 2 == 0:
        model.test()
        precision = test(model, test_image, test_label, 128)

        test_precision.append(precision)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# --- 1. Data Loading and Preprocessing ---

# Load dataset
train = pd.read_csv("../data/mnist_train.csv")
test = pd.read_csv("../data/mnist_test.csv")

# Reestructuring the data
train_label, train_image = train[train.columns[0]].to_numpy(), train[train.columns[1:]].to_numpy().reshape((-1, 1, 28, 28)) / 255
test_label, test_image = test[test.columns[0]].to_numpy(), test[test.columns[1:]].to_numpy().reshape((-1, 1, 28, 28)) / 255

# One-hot encoding for labels
one_hot = np.zeros((len(train_label), 10))
for label in range(train_label.shape[0]):
    one_hot[label, train_label[label]] = 1
train_label = one_hot

one_hot = np.zeros((len(test_label), 10))
for label in range(test_label.shape[0]):
    one_hot[label, test_label[label]] = 1
test_label = one_hot


# --- 2. Custom Dataset Class ---

class EMNISTDataset(Dataset):
    def __init__(self, image_data, labels):
        self.image_data = torch.Tensor(image_data)
        self.labels = torch.Tensor(labels)
    
    def __len__(self):
        return len(self.image_data)
    
    def __getitem__(self, idx):
        return self.image_data[idx], self.labels[idx]

# Convert data to torch tensors and create DataLoaders
train_dataset = EMNISTDataset(train_image, train_label)
test_dataset = EMNISTDataset(test_image, test_label)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)


# --- 3. Define Model ---

class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.conv1 = nn.Conv2d(1, 6, 5, 1, bias=False)
        self.relu1 = nn.ReLU()
        self.maxpool1 = nn.MaxPool2d(2, 2)  # Padding is set to 0

        self.conv2 = nn.Conv2d(6, 16, 5, 1, bias=False)
        self.maxpool2 = nn.MaxPool2d(2, 2)  # Padding is set to 0

        self.bn1 = nn.BatchNorm1d(256)  
        self.lin1 = nn.Linear(256, 84, bias=False)
        self.drop1 = nn.Dropout(0.4)
        self.bn2 = nn.BatchNorm1d(84)
        self.drop2 = nn.Dropout(0.2)
        self.lin2 = nn.Linear(84, 47, bias=False)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.maxpool1(x)
        x = self.conv2(x)
        x = self.maxpool2(x)
        x = x.view(-1, 256)  # Flatten the tensor
        x = self.bn1(x)
        x = self.lin1(x)
        x = self.drop1(x)
        x = self.bn2(x)
        x = self.drop2(x)
        x = self.lin2(x)
        return x


# --- 4. Training Function ---

def train(model, train_loader, loss_fn, optim):
    model.train()
    total_loss = 0
    total_correct = 0
    total_samples = 0

    for inputs, labels in train_loader:
        optim.zero_grad()

        # Forward pass
        outputs = model(inputs)
        loss = loss_fn(outputs, labels.argmax(dim=1))  # Use argmax for multi-class labels

        # Backward pass
        loss.backward()
        optim.step()

        # Calculate accuracy
        _, predicted = torch.max(outputs.data, 1)
        total_samples += labels.size(0)
        total_correct += (predicted == labels.argmax(dim=1)).sum().item()

        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    avg_accuracy = total_correct / total_samples

    print(f"Train Loss: {avg_loss:.4f}, Train Accuracy: {avg_accuracy * 100:.2f}%")
    return avg_loss, avg_accuracy


# --- 5. Test Function ---

def test(model, test_loader):
    model.eval()  # Set the model to evaluation mode
    correct = 0
    total = 0

    with torch.no_grad():  # We don't need to track gradients during testing
        for inputs, labels in test_loader:
            outputs = model(inputs)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels.argmax(dim=1)).sum().item()

    accuracy = correct / total
    print(f'Test Accuracy: {accuracy * 100:.2f}%')
    return accuracy


# --- 6. Training Loop ---

losses = []
train_precision = []
test_precision = []

epochs = 50
model = Model()
loss_fn = nn.CrossEntropyLoss()  # Use CrossEntropyLoss for multi-class classification
optim = optim.Adadelta(model.parameters())

for epoch in range(epochs):
    print(f"Epoch {epoch + 1}\n-------------------------------")
    model.train()

    loss, precision = train(model, train_loader, loss_fn, optim)  # Pass DataLoader for train

    losses.append(loss)
    train_precision.append(precision)

    if epoch % 2 == 0:
        precision = test(model, test_loader)  # Pass DataLoader for test
        test_precision.append(precision)


# --- 7. Plotting Loss and Accuracy ---

# Plot loss vs epochs
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(losses, label='Training Loss')
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# Plot accuracy vs epochs
plt.subplot(1, 2, 2)
plt.plot(train_precision, label='Training Accuracy')
plt.plot(test_precision, label='Testing Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

```

#### Creation of API

```python
import uvicorn
from fastapi import Request, FastAPI
from fastapi.middleware.cors import CORSMiddleware
import nest_asyncio

app = FastAPI()
model.eval()

app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

@app.post("/predict")
async def predict(request: Request):
    res = await request.json()
    # Convert the received data into a NumPy array and reshape it for the model
    image_data = np.array(res).reshape((1, 1, 28, 28)).astype(np.float32)
    
    # Convert NumPy array to PyTorch tensor and normalize the image
    picture = torch.tensor(image_data)  # Convert to PyTorch tensor
    picture = picture / 255.0  # Normalize the image to [0, 1]

    # Make the prediction
    with torch.no_grad():  # We don't need gradients for inference
        outputs = model(picture)  # Forward pass
        _, predicted = torch.max(outputs.data, 1)  # Get the index of the max log-probability

    # Return the predicted number as a response
    return {"prediction": int(predicted.item())}

if __name__ == "__main__":
    
    nest_asyncio.apply()  # Allow asyncio.run to work in a running loop
    uvicorn.run(app, host="127.0.0.1", port=3838)
```
